```python
# Need to Data Clean
# Missing values for every year - Why; perhaps a principle?
# Remove "Unnamed: 60"
# Sub Data Frame for Arab World
# Find rural and urban populations Indicator codes
# Find birth and death rates Indicator codes
```

```python
# Remove indexed column at end of data
health_df = health_df.drop(columns=["Unnamed: 60"])
```

```python
health_df.head(5)
```

```python
# Sub data frame for Country Name of Arab World
arab_world_df = health_df[health_df['Country Name'] == 'Arab World']
arab_world_df.head(5)
```

```python
# Indicator Codes for birth per 1000 people, death per 1000 people, urban population, rural population

aw_birth_df = arab_world_df[arab_world_df['Indicator Code'] == 'SP.DYN.CBRT.IN']
aw_death_df = arab_world_df[arab_world_df['Indicator Code'] == 'SP.DYN.CDRT.IN']
aw_urban_df = arab_world_df[arab_world_df['Indicator Code'] == 'SP.URB.TOTL']
aw_rural_df = arab_world_df[arab_world_df['Indicator Code'] == 'SP.RUR.TOTL']
```

```python
# POP percentage
aw_urban_percent_df = arab_world_df[arab_world_df['Indicator Code'] == 'SP.URB.TOTL.IN.ZS']
aw_rural_percent_df = arab_world_df[arab_world_df['Indicator Code'] == 'SP.RUR.TOTL.ZS']
```

```python
# pulled only columns for years
# then pulled all values per year columns
birth_year_df = aw_birth_df.iloc[:, 4:59]
birth_year_values = birth_year_df.iloc[0,:]

death_year_df = aw_death_df.iloc[:, 4:59]
death_year_values = death_year_df.iloc[0,:]

urban_year_df = aw_urban_df.iloc[:, 4:59]
urban_year_values = urban_year_df.iloc[0,:]

rural_year_df = aw_rural_df.iloc[:, 4:59]
rural_year_values = rural_year_df.iloc[0,:]

urban_percent_df = aw_urban_percent_df.iloc[:, 4:59]
urban_percent_values = urban_percent_df.iloc[0,:]

rural_percent_df = aw_rural_percent_df.iloc[:, 4:59]
rural_percent_values = rural_percent_df.iloc[0,:]

total_population_df = rural_year_values + urban_year_values
total_population_df
```

```python
# # Create a figure and axis
# fig, ax1 = plt.subplots()

# # Plot the first line with its y-axis
# ax1.plot(data['x'], data['y1'], 'b-', label='y1 = x^2')
# ax1.set_xlabel('X Axis')
# ax1.set_ylabel('Y1 Axis', color='b')
# ax1.tick_params(axis='y', labelcolor='b')

# # Create a second y-axis and plot the second line
# ax2 = ax1.twinx()
# ax2.plot(data['x'], data['y2'], 'r-', label='y2 = x')
# ax2.set_ylabel('Y2 Axis', color='r')
# ax2.tick_params(axis='y', labelcolor=
```

```python
# KEEP
# Urban Pop with Birth / Death per year
fig, ax1 = plt.subplots(figsize=(20,10))

ax1.plot(birth_year_values, color='green', label='Birth')
ax1.plot(death_year_values, color='red', label='Death')
ax1.set_xlabel('Year')
ax1.set_ylabel('Birth & Death Per 1000')
ax1.tick_params(axis='y', labelcolor='b')
plt.xticks(rotation=90)

ax2 = ax1.twinx()
ax2.bar(x=urban_year_values.index, height=urban_year_values, alpha=.3, label='Urban Population')
ax2.bar(x=rural_year_values.index, height=rural_year_values, alpha=.3, label='Rural Population')
ax2.set_ylabel('Urban Population')
ax2.tick_params(axis='y', labelcolor='b')
plt.title('Urban / Rural Births and Deaths per Year with Annual Population Growth in Arab World')

fig.legend(loc='upper right', title='Legend')
plt.show()
```

```python
# Urban / Rural Percentage Population Shift vs. Total Population Growth in Arab World
fig, ax1 = plt.subplots(figsize=(20,10))

ax1.plot(urban_percent_values, color='green', label='Urban Percent Values')
ax1.plot(rural_percent_values, color='red', label='Rural Percent Values')
ax1.set_xlabel('Year')
ax1.set_ylabel('Perecent of Total Population')
ax1.tick_params(axis='y', labelcolor='b')
plt.xticks(rotation=90)

ax2 = ax1.twinx()
ax2.bar(x=total_population_df.index, height=total_population_df, alpha=.3, label='Total Population')
ax2.set_ylabel('Population')
ax2.tick_params(axis='y', labelcolor='b')
plt.title('Urban / Rural Percentage Population Shift vs. Total Population Growth in Arab World')

fig.legend(loc='upper right', title='Legend')
plt.show()
```

```python
# Rural Pop with Birth / Death per year
fig, ax1 = plt.subplots(figsize=(20,10))

ax1.plot(birth_year_values, color='green', label='Birth')
ax1.plot(death_year_values, color='red', label='Death')
ax1.set_xlabel('Year')
ax1.set_ylabel('Birth & Death Per 1000')
ax1.tick_params(axis='y', labelcolor='b')
plt.xticks(rotation=90)

ax2 = ax1.twinx()
ax2.bar(x=rural_year_values.index, height=rural_year_values, alpha=.4, label='Population')
ax2.set_ylabel('Rural Population')
ax2.tick_params(axis='y', labelcolor='b')
plt.title('Rural Births and Deaths per Year with Annual Population Growth in Arab World')

fig.legend(loc='right', title='Legend')
plt.show()
```
